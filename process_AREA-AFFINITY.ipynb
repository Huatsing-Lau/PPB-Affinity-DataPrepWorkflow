```python
import pandas as pd
import numpy as np
import tqdm
```

```python
filepath1_area_affinity = r"E:\1.THPRD\1.AI_LifeScience\3.AntiBody\我们的创新\1.RDE-PPdG\参考文献\公开数据集\AREA-AFFINITY\supp\Data and results for protein-protein binding affinity prediction\1. Different areas in the structures of protein-protein complexes.xlsx"
df1_area_affinity = pd.read_excel(
    filepath1_area_affinity, sheet_name='Sets', header=0,
    usecols=[1,4]
)
df1_area_affinity.rename(columns={
    'PDB code': 'PDB',
}, inplace=True)
df1_area_affinity = df1_area_affinity[~df1_area_affinity['PDB'].isna()]
df1_area_affinity = df1_area_affinity[~df1_area_affinity['Binding Affinity (log(Ki) or log(Kd))'].isna()]
# PDBs_1_area_affinity = df1_area_affinity[~df1_area_affinity['PDB code'].isna()]['PDB code'].apply(lambda x: x.split('_')[0].upper()).tolist()




filepath2_area_affinity = r'E:\1.THPRD\1.AI_LifeScience\3.AntiBody\我们的创新\1.RDE-PPdG\参考文献\公开数据集\AREA-AFFINITY\supp\Data and results for antibody-protein antigen binding affinity prediction\1. Different areas in the structures of antibody-protein antigen complexes.xlsx'
df2_area_affinity = pd.read_excel(
    filepath2_area_affinity, sheet_name='Area-descriptors',
    usecols=[1,4]
)
df2_area_affinity.rename(columns={
    'PDB code': 'PDB',
}, inplace=True)
df2_area_affinity = df2_area_affinity[~df2_area_affinity['PDB'].isna()]
df2_area_affinity = df2_area_affinity[~df2_area_affinity['Binding Affinity (log(Ki) or log(Kd))'].isna()]
# PDBs_2_area_affinity = df2_area_affinity[~df2_area_affinity['PDB code'].isna()]['PDB code'].apply(lambda x: x.split('_')[0].upper()).tolist()


df = pd.concat((df1_area_affinity, df2_area_affinity), axis=0, ignore_index=True)

df['KD(M)'] = df['Binding Affinity (log(Ki) or log(Kd))'].apply(lambda x: 10**x)
for ind in df.index:
    pdb_str = df.loc[ind,'PDB']
    pdb, chains = pdb_str.split('_')
    pdb = pdb.upper()
    ligand, receptor = chains.split(':')
    df['Ligand Chains'] = ', '.join([chain.upper() for chain in ligand])
    df['Receptor Chains'] = ', '.join([chain.upper() for chain in receptor])
    df.loc[ind,'PDB'] = pdb.upper()

display(df)
```
